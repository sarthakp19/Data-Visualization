# Visualizing Loans Awarded by Kiva

In this project we'll visualize insights using a dataset from Kaggle.  

Using Seaborn, we'll explore the average loan amount by country using aggregated bar charts. We'll also visualize the distribution of loan amounts by project type and gender using box plots and violin plots.



```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('kiva_data.csv')
print(df.country.unique())
df.head(20)
```

# Visualization 1

A bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.

```python
# Creates the figure
f, ax = plt.subplots(figsize=(10, 8))
sns.set_palette(sns.color_palette('Set1'))
ax = sns.set_style('whitegrid')
ax = sns.barplot(data=df,x='country',y='loan_amount')

plt.show()
```

# Visualization 2

We can visualize even more data on one bar plot by visualizing the loan amount by country, and "nesting" by gender.

```python
# Creates the figure, you're only using this syntax so you can modify the y-axis ticks below
f, ax = plt.subplots(figsize=(15, 10))
sns.set_palette('Set2')
sns.set_style('whitegrid')
sns.barplot(data=df, x="country", y="loan_amount",hue='gender')
plt.show()
```

# Analysis

On average, male recipients receive larger loans from Kiva .

El Salvador has the *least* disparity in loan amounts awarded by gender.

Kiva should work to decrease the gender loan gap.

Kiva could hold workshops focused on women-led projects (to see more women-led project).

Kiva could require that an equal amount of loans be given to male and female-driven projects.

# Visualization 3

We are going to make a box plot to compare the distribution of loans by country.

```python
plt.figure(figsize=(10, 8))
sns.set_palette('Dark2')
sns.boxplot(data=df, x="country", y="loan_amount")
plt.show()

```

# Analysis

Kenya has the widest distribution.

Combodia would we be most likely to receive the largest loan amount .

# Visualization 4

Instead of visualizing the loan amount by *country*, we are ploting the loan amount by *activity*.

```python
plt.figure(figsize=(16, 10))
sns.set_palette('Accent')
sns.set_palette("Spectral")
sns.boxplot(data=df, x="activity", y="loan_amount")

```

# Analysis

It shows us the loan amounts by activity, rather than country. 

On average, "Farming" receives the largest loan amount and also has the widest distribution.

# Visualization 5

To visualize the distribution of the exact same data as a violin plot we could pass the same parameters to `sns.violinplot()` instead of `sns.boxplot()`. Loan Amount by Country.


```python
sns.set_palette("Spectral")
plt.figure(figsize=(13, 8))

sns.violinplot(data=df, x="country", y = "loan_amount", hue="gender", split=True)
plt.show()
```

# Analysis

In Pakistan, the average loan amount received by women is much less than the average loan amount received by men.

The same can be said for Cambodia (more or less).

In El Salvador, there is a fairly equal distribution of loans (total number of loans) between genders, as well as
similar median loan amounts (on average).
